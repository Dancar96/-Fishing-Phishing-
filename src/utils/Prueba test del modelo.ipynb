```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.metrics import recall_score
```

```python
data = pd.read_csv('../data/processed/dataset_modelo.csv')
data
```

```python
data_train = data.iloc[:9430, :]
data_train
```

```python
data_test = data.iloc[9430:, :]
data_test
```

```python
X_train = data_train.iloc[:, :-1]
y_train = data_train['status']
```

```python
xgb = XGBClassifier(colsample_bytree=0.5, learning_rate=0.1, max_depth=None, min_child_weight=1, n_estimators=200, reg_alpha=0.1, reg_lambda=0.1, subsample=0.9)
xgb.fit(X_train,y_train)
```

```python
prediccion_train = xgb.predict(X_train)
recall_train = recall_score(y_train, prediccion_train)

print('Recall en train:', recall_train)
```

```python
X_test = data_test.iloc[:, :-1]
y_test = data_test['status']
prediccion = xgb.predict(X_test)
```

```python
recall = recall_score(y_test, prediccion)

print('Recall en test:', recall)
```
